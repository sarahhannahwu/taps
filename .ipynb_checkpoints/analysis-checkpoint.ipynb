# Analysis Walkthrough of TAPS 
## Add libraries

```python
# Standard library imports
import os
import sys
from datetime import datetime, timedelta

# Data manipulation and analysis
import pandas as pd
import numpy as np

# Visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns
```

## Import your school's data

```python
school_data = pd.read_csv("student_synthetic_data.csv")
school_data_num = pd.read_csv("student_synthetic_data_num.csv")

# Display the first few rows of the dataframe
pd.DataFrame.head(school_data)
pd.DataFrame.head(school_data_num)

# Compute average satisfaction, compliance, and enforcement and their standard deviations
# Skip the first row, which contains metadata
school_data_labels = school_data.iloc[1:]
school_data_numbers = school_data_num.iloc[1:]
# Convert relevant columns to numeric types
school_data_numbers['pif_satisfied'] = pd.to_numeric(school_data_numbers['pif_satisfied'], errors='coerce')
school_data_numbers['pif_noncompliance'] = pd.to_numeric(school_data_numbers['pif_noncompliance'], errors='coerce')
school_data_numbers['pif_enforcement'] = pd.to_numeric(school_data_numbers['pif_enforcement'], errors='coerce')

average_satisfaction = school_data_numbers['pif_satisfied'].mean()
average_noncompliance = school_data_numbers['pif_noncompliance'].mean()
average_enforcement = school_data_numbers['pif_enforcement'].mean()

std_satisfaction = school_data_numbers['pif_satisfied'].std()
std_noncompliance = school_data_numbers['pif_noncompliance'].std()
std_enforcement = school_data_numbers['pif_enforcement'].std()

# Print the results
print(f"Average Satisfaction: {average_satisfaction:.2f}")
print(f"Average Noncompliance: {average_noncompliance:.2f}")
print(f"Average Enforcement: {average_enforcement:.2f}")
```

```python

# Create a boxplot to visualize the distributions of satisfaction, noncompliance, and enforcement
data_to_plot = [school_data_numbers['pif_satisfied'].dropna(),
                school_data_numbers['pif_noncompliance'].dropna(),
                school_data_numbers['pif_enforcement'].dropna()]
plt.boxplot(data_to_plot, labels=['Satisfaction', 'Noncompliance', 'Enforcement'])
plt.title('Distribution of Satisfaction, Noncompliance, and Enforcement')
plt.ylabel('Percentage of Students')
plt.show()
```

## Generate distribution of responses to the agreement questions

```python
# For questions policy_impact_1 through policy_impact_17, generate individual barplots to visualize the distribution of responses
# What percentage of students responded to each question with each possible answer?

policy_questions = [f'policy_impact_{i}' for i in range(1, 18)]

# Specify the order of the responses
response_order = ['Strongly disagree', 'Somewhat disagree', 'Neither agree nor disagree', 'Somewhat agree', 'Strongly agree']

# Specify what each policy question is about
policy_descriptions = {
    'policy_impact_1': 'I have focused better in class',
    'policy_impact_2': 'I have had fewer distractions',
    'policy_impact_3': 'I have developed better social skills',             
    'policy_impact_4': 'I have felt closer to my friends',
    'policy_impact_5': 'I have felt more left out or excluded',
    'policy_impact_6': 'I have felt more anxious without my phone',
    'policy_impact_7': 'I have talked to more people during lunch or breaks',
    'policy_impact_8': 'I have felt less pressure to check my phone',
    'policy_impact_9': 'I have gotten into trouble less often',
    'policy_impact_10': 'I have been involved in less social drama at school',
    'policy_impact_11': 'My grades have improved',
    'policy_impact_12': 'I have been less likely to cheat',
    'policy_impact_13': 'I have spent less time on my phone',
    'policy_impact_14': 'I have been bullied more',
    'policy_impact_15': 'I have felt less stressed',
    'policy_impact_16': 'I have used my phone more before and after school',
    'policy_impact_17': 'Nothing has really changed for me'
}
# For each question, calculate the counts and percentages of each response
for question in policy_questions:
    response_counts = school_data_labels[question].value_counts()
    question_label = policy_descriptions.get(question, question)
    response_percentages = (response_counts / response_counts.sum()) * 100

    # Create a bar plot for the current question
    plt.figure(figsize=(10, 6))
    sns.barplot(x=response_percentages.index, order=response_order,y=response_percentages.values, palette='viridis')
    plt.title(f'Distribution of Responses for {question}')
    plt.xlabel('Response')
    plt.ylabel('Percentage of Students')
    plt.ylim(0, 100)
    plt.show()
   
```
